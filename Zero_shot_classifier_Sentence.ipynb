```python
from transformers import pipeline
```

```python
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")
```

```python
# Define the candidate labels
candidate_labels = ["technology","sports","politics"]

# Define the sentence you want to classify
sentence = "And you don't see any party in mainstream politics that's got an alternative"  # Replace with your actual sentence

# Perform zero-shot classification
result = classifier(sentence, candidate_labels)

# Extract the top label and its score
top_label = result['labels'][0]
top_score = result['scores'][0]

# Print the results
print(f"Sentence: {sentence}")
print(f"Top Label: {top_label}")
print(f"Top Score: {top_score}")
```
